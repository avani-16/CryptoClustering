```python
# Import required libraries and dependencies

import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
plot = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90,
   
)
plot
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_market_data = StandardScaler().fit_transform(
    df_market_data
)
scaled_market_data[:5]
```

```python
df_market_data.index.unique
```

```python
# Create a DataFrame with the scaled data
market_data_scaled_df = pd.DataFrame(scaled_market_data, columns=df_market_data.columns)

# Copy the crypto names from the original data
market_data_scaled_df["coin_id"] = df_market_data.index

# Set the coinid column as index
market_data_scaled_df = market_data_scaled_df.set_index("coin_id")

# Display sample data
market_data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k_values:
    model = KMeans(n_clusters=i, n_init=10)
    model.fit(market_data_scaled_df)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {"k" : k_values,
              "inertia" : inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_dict)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_plot = df_elbow.hvplot.line(x = "k",
                                  y = "inertia",
                                  title = "Elbow Curve for K Values")
elbow_plot
#print(type(elbow_plot))
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** By looking at the Elbow curve above, it seems like at '4' the rate of decrease in distortion(inertia) starts to slow down significantly. So, best value for k is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters = 4, n_init = 10)
```

```python
# Fit the K-Means model using the scaled data
model.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_group = model.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
print(k_group)
```

```python
# Create a copy of the DataFrame
market_data_prediction_df = market_data_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_data_prediction_df["predicted_cluster"] = k_group

# Display sample data
market_data_prediction_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plot_crypto = market_data_prediction_df.hvplot.scatter(
    x = "price_change_percentage_24h",
    y = "price_change_percentage_7d",
    by = "predicted_cluster",
    hover_cols = "coin_id",
    title = "Scatter plot for predicted cluster - k = 4",
    frame_width = 600,
    frame_height = 300
).opts(yformatter="%.0f", xformatter="%.0f")

plot_crypto
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
#market_data_pca = pca.fit_transform(market_data_prediction_df)
market_data_pca = pca.fit_transform(market_data_scaled_df)

# View the first five rows of the DataFrame. 
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Find total explained variance of the three principal components by adding all 3 explained variance
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  About 89.50% of the total variance is condensed into the 3 PCA variables.

```python
# Create a new DataFrame with the PCA data.
crypto_pca = pd.DataFrame(
    market_data_pca,
    columns = ["PC1", "PC2", "PC3"]

)
#crypto_pca.head()

# Creating a DataFrame with the PCA data

# Copy the crypto names from the original data
crypto_pca["coin_id"] = list(market_data_prediction_df.index)
#crypto_pca.head()

# Set the coinid column as index
crypto_pca = crypto_pca.set_index("coin_id")

# Display sample data
crypto_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, n_init=10)
    k_model.fit(crypto_pca)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_pca_plot = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="PCA k values Elbow Curve", 
    xticks=k
)
elbow_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Best value for k when using PCA is 4 because the inertia slow down there and elbow shape starts there.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the k values using PCA value is same as original data. However, the curves are slightly different from each other. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, n_init=10)
```

```python
# Fit the K-Means model using the PCA data
model.fit(crypto_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
y_pca= model.predict(crypto_pca)
# Print the resulting array of cluster values.
print(y_pca)
```

```python
# Create a copy of the DataFrame with the PCA data

crypto_pca_prediction_df = crypto_pca.copy()
# Add a new column to the DataFrame with the predicted clusters
crypto_pca_prediction_df["pca_cluster"] = y_pca

# Display sample data
crypto_pca_prediction_df
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plot_pca_crypto = crypto_pca_prediction_df.hvplot.scatter(
    x = "PC1",
    y = "PC2",
    by = "pca_cluster",
    hover_cols = "coin_id",
    title = "Scatter plot for predicted cluster - k = 4",
    frame_width = 600,
    frame_height = 300
).opts(yformatter="%.0f", xformatter="%.0f")

plot_pca_crypto
```
